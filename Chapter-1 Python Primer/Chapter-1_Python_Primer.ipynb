```python
# P-1.3 Write a short Python function, minmax(data), that takes a sequence of one or more numbers,
# and returns the smallest and largest numbers, in the form of a tuple of length two. 
# Do not use the built-in functions min or max in implementing your solution.

def minmax(data):
    """
      Returns the smallest and largest numbers in the sequence `data`.

      Args:
        data: A sequence of numbers.

      Returns:
        A tuple of length two, containing the smallest and largest numbers in `data`.
    """
    # using built-in functions min and max
    # return min(data), max(data) BUT we can not use the built-in min and max function
    
    min_val = data[0]
    max_val = data[0]
    
    for num in data:
        if num < min_val:
            min_val = num
        elif num > max_val:
            max_val = num
    
    return min_val, max_val

minmax((1, 2, 3, 4, 5))
    
```

```python
# P-1.14 Write a short Python function that takes a sequence of integer values and determines 
# if there is a distinct pair of numbers in the sequence whose product is odd.

def has_odd_product(data):
    for i in range(len(data)):
        for j in range(i+1, len(data)):
            if data[i] != data[j] and (data[i]*data[j]) % 2 != 0:
                return True
    
    return False


sequence1 = [1, 2, 3, 4, 5]
print(has_odd_product(sequence1))   # the sequence [1, 2, 3, 4, 5] contains the distinct pair [1, 3] whose product is 3,
                                    # an odd number, so the output is True.

sequence2 = [2, 4, 6, 8]
print(has_odd_product(sequence2)) 
```

P(n, r) = n! / (n - r)!

where:

n is the total number of distinct elements in the set.
r is the number of elements to be selected from the set.

In our case is equivalent to calculating P(6, 6) since there are 6 elements in the list and all 6 elements are selected.

Therefore, the number of permutations generated by itertools.permutations(characters) in this case is:

P(6, 6) = 6! / (6 - 6)! = 6! / 0! = 6! / 1 = 6 * 5 * 4 * 3 * 2 * 1 = 720 permutations.

So, the code will generate and print a total of 720 permutations of the characters 'c', 'a', 't', 'd', 'o', and 'g'.

```python
# P-1.29 Write a Python program that outputs all possible strings formed by using the characters c , a , t , d , o , and g 
# exactly once.

import itertools

def generate_strings():
    
    """The program defines a function generate_strings that uses itertools.permutations to generate all possible 
    permutations of the characters 'c', 'a', 't', 'd', 'o', and 'g'.

    It loops over each permutation and contatenates the characters together to form a string.

    The resulting string is then printed.

    In the main part of the code, the generate_strings function is called to generate and print all possible strings 
    formed by the given characters.

    The program generates and outputs all possible strings that can be formed using the characters 
    'c', 'a', 't', 'd', 'o', and 'g' exactly once.
    """
    characters = ['c', 'a', 't', 'd', 'o', 'g']
    
    permutations = itertools.permutations(characters)
    
    #print(list(permutations))
    
    count = 0
    for permutation in permutations:
        string = ""
        count += 1
        for char in permutation:
            string += char
        print(string)
    print(count)


# Call the function to generate and print all possible strings
generate_strings()


```

```python
# P-1.30 Write a Python program that can take a positive integer greater than 2 as input and write out the number 
# of times one must repeatedly divide this number by 2 before getting a value less than 2.

def find_count(num):
    """Divides a number by 2 repeatedly until the value is less than 2.

      Args:
        number: The number to divide.

      Returns:
        The number of times the number was divided by 2.
      """
    count = 0
    while(num > 2):
        num = num/2
        count = count + 1
    return count


n = int(input("Enter a positive integer greater than 2: "))
print(f"Number of times to divide by 2: {find_count(n)}")
```

```python
def make_change(charged, given):
    change = given - charged

    # Define the values of bills and coins
    values = {
        '100 bill': 100,
        '50 bill': 50,
        '20 bill': 20,
        '10 bill': 10,
        '5 bill': 5,
        '1 bill': 1,
        'quarter': 0.25,
        'dime': 0.10,
        'nickel': 0.05,
        'penny': 0.01
    }

    # Initialize the count of each bill and coin to 0
    counts = {'100 bill': 0, '50 bill': 0, '20 bill': 0, '10 bill': 0, '5 bill': 0, '1 bill': 0, 
              'quarter': 0, 'dime': 0, 'nickel': 0, 'penny': 0}
    

    # Calculate the count of each bill and coin
    # change 54.5
    # name = '50 Dollar Bill' and value = 50.0

    for name, value in values.items():
        while change >= value:
            counts[name] = counts[name] + 1
            change = change - value

    return counts

# Taking input from the user
charged = float(input("Enter the amount charged: "))
given = float(input("Enter the amount given: "))

# Calculating the change
change_counts = make_change(charged, given)

# Displaying the result
print("Change:")
for name, count in change_counts.items():
    if count > 0:
        print(f"{count} {name}")
```

```python
# The birthday paradox says that the probability that two people in a room will have the same birthday is 
# more than half, provided n, the number of people in the room, is more than 23. This property is not really a paradox, 
# but many people ﬁnd it surprising. Design a Python program that can test this paradox by a series of experiments
# on randomly generated birthdays, which test this paradox for n = 5, 10, 15, 20,... , 100.

import random

def generate_birthdays(n):
    """Generate a list of n random birthdays."""
    birthdays = []
    for _ in range(n):
        # Generate a random day of the year (1-365)
        birthday = random.randint(1, 365)
        birthdays.append(birthday)
    return birthdays

def has_duplicate(birthdays):
    """Check if there are any duplicate birthdays in the list."""
    return len(birthdays) != len(set(birthdays))

def birthday_paradox_experiment(n, num_experiments):
    """Run the birthday paradox experiment for a given number of people."""
    num_successes = 0
    for _ in range(num_experiments):
        birthdays = generate_birthdays(n)
        if has_duplicate(birthdays):
            num_successes += 1
    probability = num_successes / num_experiments
    return probability

# Run experiments for n = 5, 10, 15, ..., 100
for n in range(5, 101, 5):
    probability = birthday_paradox_experiment(n, 10000)
    print(f"For n = {n}, probability of having the same birthday: {probability}")
```

```python
# P-1.36 Write a Python program that inputs a list of words, separated by white- space, and outputs how many times each 
# word appears in the list. You need not worry about efﬁciency at this point.

def count_word_frequency(words):
    # Split the input string into a list of words
    word_list = words.split()

    # Create an empty dictionary to store word frequencies
    word_freq = {}

    # Count the frequency of each word
    for word in word_list:
        if word in word_freq:
            word_freq[word] = word_freq[word] + 1
        else:
            word_freq[word] = 1

    # Print the word frequencies
    for word, freq in word_freq.items():
        print(f"{word}: {freq}")

# Take input from the user
input_words = input("Enter a list of words separated by whitespace: ")

# Call the function to count word frequencies
count_word_frequency(input_words)
```
